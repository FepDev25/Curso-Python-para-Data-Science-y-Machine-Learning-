```python
def ver_imagen_en_fecha(fecha):
    params = {
        "api_key": api_key,
        "date": fecha
    }
    respuesta = requests.get(url, params=params)
    data = respuesta.json()
    print(data["title"])
    print(data["explanation"])
    if 'url' in data and data['media_type'] == 'image':
        display(Image(data['url']))
```

```python
url_2 = "https://api.nasa.gov/DONKI/IPS?"

def obtener_donki(fecha_inicio, fecha_fin):
    params = {
        "api_key": api_key,
        "startDate": fecha_inicio,
        "endDate": fecha_fin
    }
    respuesta = requests.get(url_2, params=params)
    data = respuesta.json()
    return pd.DataFrame(data)
```
