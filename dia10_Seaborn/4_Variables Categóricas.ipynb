## Categorias

- Visualización de datos categóricos

```python
import seaborn as sns
```

```python
propinas = sns.load_dataset("tips")
propinas.head()
```

```python
sns.catplot(
    data=propinas,
    kind="strip",
    x = "day",
    y= "total_bill",
    hue="smoker"
);
```

```python
sns.catplot(
    data=propinas,
    kind="swarm",
    x = "day",
    y= "total_bill",
    hue="smoker"
);
```

- Si el conjunto es grande y se desea una representación general rápida se recomienda uso de `stripplot`.
- Si es importante visualizar la distribución de los datos se recomienda usar `swarmplot`.

```python
sns.catplot(
    data=propinas,
    kind="box",
    x = "day",
    y= "total_bill",
    hue="smoker"
);
```

- Cada caja representa la distribución de total_bill para un día determinado, separada por fumadores y no fumadores.
- La línea dentro de la caja es la mediana.
- Los bordes de la caja indican el primer (Q1) y tercer cuartil (Q3).
- Las "bigotes" (líneas horizontales que salen de las cajas) indican el rango de datos no considerados outliers.
- Los círculos fuera de las cajas representan valores atípicos (outliers).

```python
sns.catplot(
    data=propinas,
    kind="violin",
    x = "day",
    y= "total_bill",
    hue="smoker"
);
```
