```python
import seaborn as sns
```

```python
pinguinos = sns.load_dataset("penguins")
pinguinos = pinguinos.rename(columns={
    "species": "especie",
    "island": "isla",
    "bill_length_mm": "largo_pico_mm",
    "bill_depth_mm": "profundidad_pico_mm",
    "flipper_length_mm": "largo_aleta_mm",
    "body_mass_g": "masa_corporal_g",
    "sex": "sexo"
})
```

```python
g = sns.PairGrid(
    pinguinos,
    hue="especie",
) 
```

```python
g = sns.PairGrid(
    pinguinos,
    hue="especie",
    corner=True # Evitar duplicados mostrando solo los datos de la esquina inferior izquierda
) 
```

```python
g = sns.PairGrid(
    pinguinos,
    hue="especie",
    corner=True,
) 

g.map_lower(
    sns.kdeplot,
    levels=5,
    hue=None,
    color=".2",
)
```

```python
g = sns.PairGrid(
    pinguinos,
    hue="especie",
    corner=True,
) 

g.map_lower(
    sns.kdeplot,
    levels=5,
    hue=None,
    color=".2",
)

g.map_lower(
    sns.scatterplot,
    marker="+",
)
```

```python
g = sns.PairGrid(
    pinguinos,
    hue="especie",
    corner=True,
) 

g.map_lower(
    sns.kdeplot,
    levels=5,
    hue=None,
    color=".2",
)

g.map_lower(
    sns.scatterplot,
    marker="+",
)

g.map_diag(
    sns.histplot,
    element="step",
    linewidth=0,
    kde=True
)
```

```python
g = sns.PairGrid(
    pinguinos,
    hue="especie",
    corner=True,
) 

g.map_lower(
    sns.kdeplot,
    levels=5,
    hue=None,
    color=".2",
)

g.map_lower(
    sns.scatterplot,
    marker="+",
)

g.map_diag(
    sns.histplot,
    element="step",
    linewidth=0,
    kde=True
)

g.add_legend(
    frameon=True
)
```

```python
g = sns.PairGrid(
    pinguinos,
    hue="especie",
    corner=True,
) 

g.map_lower(
    sns.kdeplot,
    levels=5,
    hue=None,
    color=".2",
)

g.map_lower(
    sns.scatterplot,
    marker="+",
)

g.map_diag(
    sns.histplot,
    element="step",
    linewidth=0,
    kde=True
)

g.add_legend(
    frameon=True
)
g.legend.set_bbox_to_anchor((.61, .6))
```
