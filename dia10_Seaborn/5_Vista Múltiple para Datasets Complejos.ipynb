```python
import seaborn as sns
```

```python
pinguinos = sns.load_dataset("penguins")
pinguinos = pinguinos.rename(columns={
    "species": "especie",
    "island": "isla",
    "bill_length_mm": "largo_pico_mm",
    "bill_depth_mm": "profundidad_pico_mm",
    "flipper_length_mm": "largo_aleta_mm",
    "body_mass_g": "masa_corporal_g",
    "sex": "sexo"
})
pinguinos.head()
```

```python
sns.jointplot(
    data=pinguinos,
    x="largo_aleta_mm",
    y="largo_pico_mm",
    hue="especie"
)
```

```python
sns.pairplot(
    data=pinguinos,
    hue="especie"
)
```
