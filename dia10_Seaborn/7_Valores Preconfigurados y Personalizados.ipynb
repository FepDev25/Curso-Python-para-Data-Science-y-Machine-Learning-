```python
import seaborn as sns
import matplotlib as plt
```

```python
pinguinos = sns.load_dataset("penguins")
pinguinos = pinguinos.rename(columns={
    "species": "especie",
    "island": "isla",
    "bill_length_mm": "largo_pico_mm",
    "bill_depth_mm": "profundidad_pico_mm",
    "flipper_length_mm": "largo_aleta_mm",
    "body_mass_g": "masa_corporal_g",
    "sex": "sexo"
})
```

```python
sns.relplot(data=pinguinos,
            x="largo_pico_mm",
            y="profundidad_pico_mm",
            hue="masa_corporal_g");
```

```python
# Aplica el estilo "dark" al resto de mis grficos
sns.set_theme(style="dark")

sns.relplot(data=pinguinos,
            x="largo_pico_mm",
            y="profundidad_pico_mm",
            hue="masa_corporal_g");
```

```python
sns.set_theme(style="dark")

sns.relplot(data=pinguinos,
            x="largo_pico_mm",
            y="profundidad_pico_mm",
            hue="masa_corporal_g",
            palette="crest", # Cambia la paleta de colores
            marker="o", # Cambia el estilo de los marcadores
            s=30 # Cambia el tamao de los marcadores
);
```

```python
g = sns.relplot(data=pinguinos,
            x="largo_pico_mm",
            y="profundidad_pico_mm",
            hue="masa_corporal_g",
            palette="crest",
            marker="o", 
            s=30
);

g.set_axis_labels("Largo del pico (mm)", 
                  "Profundidad del pico (mm)",
                  labelpad=10
)
```

```python
g = sns.relplot(data=pinguinos,
            x="largo_pico_mm",
            y="profundidad_pico_mm",
            hue="masa_corporal_g",
            palette="crest",
            marker="o", 
            s=50
);

g.set_axis_labels("Largo del pico (mm)", 
                  "Profundidad del pico (mm)",
                  labelpad=10
)

g.legend.set_title("Masa corporal (g)")
```

```python
g = sns.relplot(data=pinguinos,
            x="largo_pico_mm",
            y="profundidad_pico_mm",
            hue="masa_corporal_g",
            palette="crest",
            marker="o", 
            s=50
);

g.set_axis_labels("Largo del pico (mm)", 
                  "Profundidad del pico (mm)",
                  labelpad=10
)

g.legend.set_title("Masa corporal (g)")
g.figure.set_size_inches(7, 5) # Cambia el tamao de la figura
```
