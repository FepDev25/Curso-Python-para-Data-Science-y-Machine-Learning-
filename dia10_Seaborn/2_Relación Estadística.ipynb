```python
import seaborn as sns
```

## Relación estadística: Vínculo, relación entre variables

```python
propinas = sns.load_dataset("tips")
propinas.head()
```

```python
sns.relplot(
    data=propinas,
    x = "total_bill",
    y = "tip"
)
```

```python
sns.relplot(
    data=propinas,
    x = "total_bill",
    y = "tip",
    kind="line"
)
```

```python
sns.relplot(
    data=propinas,
    x = "total_bill",
    y = "tip",
    hue="smoker"
)
```

```python
sns.relplot(
    data=propinas,
    x = "total_bill",
    y = "tip",
    hue="smoker",
    style="smoker"
)
```

```python
sns.relplot(
    data=propinas,
    x = "total_bill",
    y = "tip",
    hue="smoker",
    style="smoker",
    size="size"
)
```

```python
sns.relplot(
    data=propinas,          # El DataFrame de entrada. Aquí es un conjunto de datos llamado "propinas".
    x="total_bill",         # Eje X del gráfico: representa el total de la cuenta.
    y="tip",                # Eje Y del gráfico: representa la propina dada.
    hue="smoker",           # Color: diferencia los puntos por si la persona es fumadora o no.
    style="smoker",         # Estilo del marcador (círculo, triángulo, etc.) también por si es fumador o no.
    size="size",            # Tamaño del marcador según el número de personas en la mesa.
    col="time"              # Crea una columna de gráficos separados según el momento del día ("Lunch" o "Dinner").
)
```
