## üñºÔ∏è Nivel de Figura (`Figure-level`)

- **Funciones**: `relplot()`, `catplot()`, `displot()`, `lmplot()`, `jointplot()`
- **Devuelven**: un objeto `FacetGrid` o `JointGrid`, que **puede contener m√∫ltiples gr√°ficos (subplots)**.
- **Ventaja**: √∫tiles para **gr√°ficos complejos**, como paneles divididos por categor√≠as (`col`, `row`).
- **Limitaci√≥n**: no puedes usar directamente `ax=`, ya que Seaborn crea la figura por ti.

### Ejemplo:
```python
sns.catplot(data=df, x="tipo", y="valor", col="grupo", kind="box")
```
Esto crea una figura con varios subgr√°ficos, uno por grupo.

---

## üìä Nivel de Ejes (`Axes-level`)

- **Funciones**: `boxplot()`, `barplot()`, `scatterplot()`, `histplot()`, `violinplot()`, `heatmap()`
- **Devuelven**: un objeto `matplotlib.axes.Axes`, que es un **solo gr√°fico individual**.
- **Ventaja**: puedes integrarlo f√°cilmente en **figuras personalizadas** con `matplotlib`.
- **Uso com√∫n**: cuando necesitas **m√°s control sobre un solo gr√°fico**.

### Ejemplo:
```python
fig, ax = plt.subplots()
sns.boxplot(data=df, x="tipo", y="valor", ax=ax)
```

---

## üß† Resumen

| Caracter√≠stica         | Nivel Figura (`relplot`, `catplot`, etc.) | Nivel Ejes (`boxplot`, `barplot`, etc.) |
|------------------------|--------------------------------------------|-----------------------------------------|
| Devuelve               | `FacetGrid` / `JointGrid`                  | `Axes`                                  |
| Control m√∫ltiple       | ‚úî (columnas, filas, tama√±os)               | ‚úò (solo 1 eje)                           |
| Personalizaci√≥n con `matplotlib` | Limitada                          | Completa                                 |
| Par√°metro `ax=`        | ‚ùå No se usa                               | ‚úî Se usa                                |


````python
import seaborn as sns
````

````python
pinguinos = sns.load_dataset("penguins")
pinguinos = pinguinos.rename(columns={
    "species": "especie",
    "island": "isla",
    "bill_length_mm": "largo_pico_mm",
    "bill_depth_mm": "profundidad_pico_mm",
    "flipper_length_mm": "largo_aleta_mm",
    "body_mass_g": "masa_corporal_g",
    "sex": "sexo"
})
````

````python
sns.displot(
    data=pinguinos,
    x="largo_aleta_mm",
    hue="especie",
    multiple="stack"
)
````

````python
# Bajo nivel
sns.histplot(
    data=pinguinos,
    x="largo_aleta_mm",
    hue="especie",
    multiple="stack"
)
````

````python
sns.displot(
    data=pinguinos,
    x="largo_aleta_mm",
    hue="especie",
    col="especie"
)
````

````python
sns.relplot(
    data=pinguinos,
    x="largo_aleta_mm",
    y="largo_pico_mm",
    col="sexo"
)
````

````python
graf = sns.relplot(
    data=pinguinos,
    x="largo_aleta_mm",
    y="largo_pico_mm",
    col="sexo"
)

graf.set_axis_labels("Largo de la aleta (mm)", "Largo del pico (mm)")
````
