```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("./Iluminacion.csv")
df
```

```python
fig, ax = plt.subplots()

hb = ax.hexbin( df["Latitud"], 
        df["Longitud"], 
        C=df["Luminosidad"], 
        gridsize=50,
        cmap="Blues",
)
ax.set_facecolor("black")
plt.colorbar(hb, label="Nivel de sombras")
plt.xlabel("Latitud")
plt.ylabel("Longitud")
plt.title("Luminosidad en Ciudad GÃ³tica")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x =np.array(list(range(1, 11)))
y = x**2

fig, ax = plt.subplots(figsize=(6, 4))
ax.stem(x, y)

plt.title("RelaciÃ³n entre NÃºmeros y sus Cuadrados")
plt.xlabel("NÃºmero (x)")
plt.ylabel("Cuadrado del nÃºmero (x^2)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4,5]
y1 = [6, 1, 4, 4, 8]
y2 = [4, 6, 3, 5, 8]
y3 = [7, 9, 9, 2, 7]
y = np.vstack([y1, y2, y3])

fig, ax = plt.subplots(figsize=(6,4))
ax.stackplot(x, 
             y,
            labels=["Serie y1", "Serie y2", "Serie y3"],
            colors=["red", "blue", "green"],
)

plt.title("DistribuciÃ³n acumulativa de valores por serie de tiempo")
plt.xlabel("Tiempo")
plt.ylabel("Valores Acumulados")
plt.legend(loc="upper left")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Generar los datos para cada muestra usando numpy.random.normal
np.random.seed(0) # Establecer la misma semilla para la generaciÃ³n de datos
data1 = np.random.normal(loc=0, scale=0.5, size=100)
data2 = np.random.normal(loc=2, scale=1, size=100)
data3 = np.random.normal(loc=-2, scale=1.5, size=100)
data = [data1, data2, data3]
# plot:
fig, ax = plt.subplots(figsize=(6, 4))

ax.violinplot(data)
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['Muestra 1', 'Muestra 2', 'Muestra 3'])

plt.title("ComparaciÃ³n de distribuciones en Violin Plot")
plt.show()
```
