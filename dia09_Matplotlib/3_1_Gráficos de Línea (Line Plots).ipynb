```python
import pandas as pd
import matplotlib.pyplot as plt
import random 
import pandas as pd
%matplotlib inline
```

```python
def generar_lista_aleatoria(n):
    """
    Genera una lista de n números aleatorios entre 1 y 100.
    """
    return [random.randint(1, 100) for _ in range(n)]
```

```python
fig = plt.figure()
ax = plt.axes()
```

- grid: Son las líneas de fondo (horizontal y vertical) que ayudan a leer mejor los valores en el gráfico.

```python
plt.grid()
plt.show()
```

```python
x = list(range(1, 16))
y = generar_lista_aleatoria(15)

print(y)
plt.plot(x,y)
plt.grid()
plt.show()
```

- format: Permite darle formato a datos, como alinear texto, colocar decimales en números, agregar ceros a la izquierda, entre otros, de manera rápida y controlada.

```python
plt.plot(x,y, "ro-")
plt.grid()
plt.show()
```

```python
plt.plot(x,y, "b*:")
plt.grid()
plt.show()
```

```python
plt.plot(x,y, "mD--")
plt.grid()
plt.show()
```

```python
plt.plot(x,y)
plt.grid()
plt.axis([0, max(x) + 1, 0, max(y) + 1])
plt.show()
```
