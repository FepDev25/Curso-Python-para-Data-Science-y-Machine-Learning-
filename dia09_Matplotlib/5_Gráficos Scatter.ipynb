```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

### Gráficos Scatter
- Tomar los registros y desparramarlos como puntos en una cuadrícula.

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
```

```python
plt.plot(x,y, "o")
plt.grid()
plt.show()
```

```python
aleatorio = np.random.RandomState(0)
marcadores = ['o', 'x', '+', 's', 'D']

for marcador in marcadores:
    x = aleatorio.rand(10)
    y = aleatorio.rand(10)
    plt.plot(x, y, marcador)
plt.show()
```

```python
plt.plot(x,y, "b-o")
plt.show()
```

```python
plt.plot(x,y, "r-D", markersize=10)
plt.show()
```

```python
plt.plot(x,y, "r-D", markersize=10, linewidth=4)
plt.show()
```

```python
plt.plot(x,y, "r-D", markersize=10, linewidth=4, markerfacecolor="white")
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

```python
aleatorio = np.random.RandomState(0)
plt.scatter(aleatorio.rand(100),
            aleatorio.rand(100),
            c=aleatorio.rand(100),
            s=1000*aleatorio.rand(100),
            alpha=0.5,
)
plt.colorbar()
plt.show()
```

### Ejercicios

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y, color="r", s=50)
plt.show()
```

```python
import matplotlib.pyplot as plt

x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]

plt.scatter(x1, y1, marker="o", color="blue", label="Serie 1")
plt.scatter(x2, y2, marker="^", color="green", label="Serie 2")
plt.legend()
plt.show()
```
