```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "Nombre" :["Juan", "Pedro", "Maria"],
    "Edad" : [23, 45, 34]   
})
df1
```

```python
df2 = pd.DataFrame({
    "Nombre" :["Francisco", "Felipe", "Sandra"],
    "Edad" : [31, 13, 12]
})
df2
```

## Concatenar dataframes

```python
df_concatenado = pd.concat([df1, df2]) # Por defecto
df_concatenado
```

```python
df_concatenado = pd.concat([df1, df2], axis=1) # Axis 1 se refiere a las columnas
df_concatenado
```

```python
df_concatenado = pd.concat([df1, df2], ignore_index=True) # Indices reseteados
df_concatenado
```

```python
df_concatenado = pd.concat([df1, df2], keys=["df1", "df2"]) # Claves para identificar los dataframes
df_concatenado
```

```python
df1 = pd.DataFrame({
    'Gastos': [100, 200, 300],
    'Ventas': [400, 500, 600]
})

df2 = pd.DataFrame({
    'Gastos': [150, 250, 350],
    'Ventas': [450, 550, 650]
})

df3 = pd.DataFrame({
    'Gastos': [200, 300, 400],
    'Ventas': [500, 600, 700]
})

df_concatenado = pd.concat([df1, df2, df3], axis=0, keys=['df1', 'df2', 'df3'])
df_concatenado
```

### Ejercicios

```python
import pandas as pd


ventas_enero = pd.DataFrame({'Producto': ["Manzanas", "Bananas", "Naranjas"], 'Cantidad': [300, 200, 150]})

ventas_febrero = pd.DataFrame({'Producto': ["Manzanas", "Bananas", "Naranjas"], 'Cantidad': [350, 210, 170]})

ventas_total = pd.concat([ventas_enero, ventas_febrero], ignore_index=True)
ventas_total
```

```python
import pandas as pd

datos_cliente = pd.DataFrame({'Nombre': ["Ana", "Luis", "Marta"], 'Edad': [34, 45, 28]})
compras_cliente = pd.DataFrame({'Producto': ["Libro", "Lápiz", "Cuaderno"], 'Precio': [15.50, 0.50, 2.00]})

info_cliente = pd.concat([datos_cliente, compras_cliente], axis=1)
info_cliente
```

```python
import pandas as pd


tienda_a = pd.DataFrame({'Producto': ["Manzanas", "Bananas"], 'Cantidad': [500, 300]})
tienda_b = pd.DataFrame({'Producto': ["Naranjas", "Peras"], 'Cantidad': [400, 250]})

ventas_tienda = pd.concat([tienda_a, tienda_b], keys=['Tienda A', 'Tienda B'])
ventas_tienda
```
