```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

## 1. Cargar los datos

```python
iris = load_iris()
```

```python
def head(data, valor_head=5):
    print(data[:valor_head])

head(iris.data)
```

```python
print(iris.target)
```

```python
X = iris.data
y = iris.target
```

## 2. Estandarizar los datos

```python
x_centrado = X - np.mean(X, axis=0)
head(x_centrado)
```

## 3. PCA

```python
pca = PCA(n_components=2)
```

```python
x_pca = pca.fit_transform(x_centrado) # Aplicar la transformacioÃÅn y obtener los 2 componentes
head(x_pca)
```

```python
especies = ['setosa', 'versicolor', 'virginica']
```

## 4. Graficar

```python
plt.figure(figsize=(8, 6))
for i in range(0, 3):
    plt.scatter(
        x_pca[y == i, 0], 
        x_pca[y == i, 1], 
        label=especies[i]
    )
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend()
plt.title("PCA de Iris")
plt.show()
```

### Conclusiones del modelo PCA en este ejemplo:

- Reducci√≥n efectiva de la dimensi√≥n:

    - Pasamos de 4 caracter√≠sticas a 2, sin perder demasiada informaci√≥n (PCA conserva la m√°xima varianza posible).

- Visualizaci√≥n clara:

    - Las especies setosa, versicolor y virginica se pueden visualizar claramente en el plano 2D, especialmente setosa, que est√° bien separada.

- Separabilidad:

    - Aunque hay algo de superposici√≥n entre versicolor y virginica, PCA permite una visualizaci√≥n bastante √∫til para explorar y entender los datos.

- Aplicaci√≥n pr√°ctica:

    - Este tipo de an√°lisis es ideal para preprocesamiento en modelos de machine learning, para mejorar rendimiento y reducir complejidad.

### Ejercicios

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

iris = load_iris()  
X = iris.data
y = iris.target

X_centrado = X - np.mean(X, axis=0)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_centrado)

especies = ['setosa', 'versicolor', 'virginica']

plt.figure(figsize=(8, 6))
for i in range(0, 3):
    plt.scatter(
        X_pca[y == i, 0], 
        X_pca[y == i, 1], 
        label=especies[i]
    )
plt.xlabel('Primer componente principal')
plt.ylabel('Segundo componente principal')
plt.legend()
plt.title("PCA del conjunto Iris")
plt.show()
```
