```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
# Especios existentes
df["species"].unique()
```

### 1. Separa las variables predictoras de la variable objetivo

```python
X = df.drop("species", axis=1)
X.head()
```

### 2. Codificar variables categóricas

```python
le = LabelEncoder()
especies = le.fit_transform(df["species"])
y = especies
y
```

### 3. Separa los conjuntos de entrenamiento y prueba

```python
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X, y, train_size=0.8, random_state=42)
```

### 4. Entrena el modelo

```python
arbol = DecisionTreeClassifier()
```

```python
arbol.fit(X_entrena, y_entrena)
```

```python
plt.figure(figsize=(20, 20))
plot_tree(decision_tree=arbol)
```

```python
plt.figure(figsize=(20, 20))
plot_tree(
    decision_tree=arbol,
    filled=True
)
```

```python
plt.figure(figsize=(20, 20))
plot_tree(
    decision_tree=arbol,
    filled=True,
    class_names=["setosa", "versicolor", "virginica"]
)
```

```python
plt.figure(figsize=(20, 20))
plot_tree(
    decision_tree=arbol,
    filled=True,
    class_names=["setosa", "versicolor", "virginica"],
    feature_names=df.columns.to_list()
)
plt.savefig("arbol_iris.png")
```

- El índice Gini mide la impureza de los datos en el nodo.
  - Un valor de 0.0 significa que todas las muestras del nodo pertenecen a una misma clase.
  - Valores más altos indican mayor mezcla entre clases.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree

iris_df = sns.load_dataset("iris")

X = iris_df.drop("species", axis=1)

le = LabelEncoder()
especies = le.fit_transform(iris_df["species"])
y = especies

X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X, y, train_size=0.8, random_state=42)

arbol = DecisionTreeClassifier()
arbol.fit(X_entrena, y_entrena)

plt.figure(figsize=(20, 20))
plot_tree(decision_tree=arbol,
        filled = True,
        class_names=["setosa", "versicolor", "virginica"],
        feature_names = iris_df.columns.to_list());
```
