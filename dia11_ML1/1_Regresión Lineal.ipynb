```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.DataFrame()
df["Area"] = [2600,3000,3200,3600,4000]
df["Precio"] = [550000,565000,610000,680000,725000]
```

```python
plt.plot(
    df["Area"],
    df["Precio"],
    "bo-"
)
plt.show()
```

## 1. Entrenar el algoritmo

```python
# Variable independiente
X = df[["Area"]] # Doble corchete porque necesitamos un dataframe
X
```

```python
# Variable dependiente
Y = df["Precio"]
Y
```

```python
modelo = linear_model.LinearRegression()
```

```python
modelo.fit(X,Y)
```

## 2. Hacer predicciones

```python
modelo.predict([[3350]])
```

```python
plt.scatter(
    df["Area"],
    df["Precio"],
)
plt.plot(
    df["Area"],
    modelo.predict(X),
    "r-"
)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.DataFrame()
df["Edad"] =  [25, 30, 35, 40, 45]
df["Salario"] =  [3000, 3500, 4000, 4500, 5000]

X = df[["Edad"]]
Y = df["Salario"]

modelo = linear_model.LinearRegression()
modelo.fit(X,Y)

plt.scatter(
    df["Edad"],
    df["Salario"],
    color="blue",
    marker="o"
)
plt.plot(
    df["Edad"],
    modelo.predict(X),
)
plt.title('Relación entre Edad y Salario')
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.grid(True)
plt.show()
```
