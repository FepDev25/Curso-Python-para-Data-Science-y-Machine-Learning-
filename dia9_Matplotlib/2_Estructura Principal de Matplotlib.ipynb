```python
import matplotlib.pyplot as plt
```

```python
numeros = [4,2,7,6,3]

plt.plot(numeros)
plt.show()
```

- Figure: Es la ventana o el lienzo general donde se dibujan las gráficas. Es como la "hoja" completa.
- Axes: Es cada gráfica individual dentro del figure, donde realmente se dibujan los datos (con sus propios ejes x, y, etc.).

```python
numeros = [4,2,7,6,3]

plt.figure(facecolor="lightblue")
plt.plot(numeros);
plt.show()
```

```python
plt.subplots(ncols=2, nrows=2) # 4 subplots
```

```python
fig, axs = plt.subplots(ncols=2, nrows=2)
fig
```

```python
axs
```

```python
print(type(fig))
print(type(axs))
print(type(axs[0][0]))
```

```python
axs1 = axs[0][0]
axs1.plot(numeros)

axs2 = axs[0][1]
axs2.plot([1,2,3,4,5])

axs3 = axs[1][0]
axs3.plot([5,4,3,2,1])

axs4 = axs[1][1]
axs4.plot([10,50,10,50,100])
```
