```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
df = pd.DataFrame({
    "etiquetas":['Manzanas', 'Bananas', 'Naranjas', 'Uvas'],
    "cantidad":[30, 60, 80, 20]
})
df
```

```python
plt.pie( df["cantidad"],
        labels=df["etiquetas"], # Etiquetas
        autopct='%1.1f%%', # Porcentajes
        startangle=90, # Comenzar desde el Ã¡ngulo 90
)

plt.show()
```

```python
plt.pie( df["cantidad"],
        labels=df["etiquetas"], # Etiquetas
        autopct='%1.1f%%', # Porcentajes
        startangle=90, # Comenzar desde el Ã¡ngulo 90
        shadow=True, # Sombra
        explode=(0.1, 0, 0, 0), # Separar la primera porciÃ³n
        colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'], # Colores
        radius=1.2, # Radio del cÃ­rculo
        counterclock=False, # Sentido de las agujas del reloj
)

plt.show()
```

```python
plt.pie( df["cantidad"],
        labels=df["etiquetas"], 
        autopct='%1.1f%%',
        startangle=90,
        explode=(0, 0, 0.1, 0), 
        colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'], 
        radius=2, 
        counterclock=False,
)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    "frutas": ['Manzanas', 'Bananas', 'Cerezas', 'Duraznos'],
    "cantidades": [40, 25, 15, 20]
})

plt.pie(df["cantidades"],
        labels=df["frutas"],
        autopct='%1.1f%%',
        colors = ['red', 'yellow', 'darkred', 'orange'],
        shadow=True,
        explode=(0, 0, 0.1, 0), 
)
plt.title("DistribuciÃ³n de frutas en el almacÃ©n")
plt.show()
```
