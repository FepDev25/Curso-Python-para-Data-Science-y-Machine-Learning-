```python
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
fig, axs = plt.subplots(ncols=2, nrows=1)
fig
```

```python
fig, axs = plt.subplots(ncols=2, nrows=1, figsize=(10, 5))
ax1 = axs[0]
ax2 = axs[1]
```

```python
ax1
```

```python
ax2
```

```python
ax1.plot(list(range(1,16)), random.sample(range(1, 100), 15))
ax1.set_title('Gráfico de línea')

ax2.scatter(list(range(1,16)), random.sample(range(1, 100), 15))
ax2.set_title('Gráfico Scatter')
fig
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(ncols=2, nrows=1)

axs[0].plot( [1, 2, 3, 4], [2, 4, 8, 16], c='green') 
axs[0].set_title('Gráfico de línea')

axs[1].scatter( [1, 2, 3, 4], [2, 4, 8, 16], c='red')
axs[1].set_title('Gráfico de scatter')

fig
```
